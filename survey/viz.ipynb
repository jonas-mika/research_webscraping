# Viz 1
Digitality vs. Nondigitality

```python
#categories = ["Digital", "Nondigital"]
counts = [93, 7]

fig = plt.figure(figsize=(8,3))

axes = fig.add_axes([0.15,0.15,0.7,0.7])

labels = 'At least one digital notetaking device', 'Purely analog notes'
sizes = [93, 7]
explode = (0.1, 0.0) 
colors = ["#989bf8", "grey"]

axes.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
axes.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.savefig("digitality.pdf")
```
